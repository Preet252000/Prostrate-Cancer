```python
# reviewd from a youtube video https://youtu.be/DIxxz_DvqLA?si=pmXXpmDOjRYXpUGH
text1 = 'mean radius is 16.8 '
text2 = 'mean texture is 18.2'
compare_list = [text1, text2]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

tf_idf_matrix = tfidf_vectorizer.fit_transform(compare_list)
```

```python
tfidf_vectorizer.vocabulary_
```

```python
print (tf_idf_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = cosine_similarity(tf_idf_matrix)
print(f'The cosine similarity between two tweets is {cos_sim[0,1]}')
```
